```python
import pandas as pd

# Create the data
data = {
    'Name': ['John Doe', 'Jane Smith', 'Emily Jones', 'Chris Lee'],
    'Age': [28, 34, 24, 40],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
    'Occupation': ['Software Engineer', 'Data Scientist', 'Graphic Designer', 'Project Manager']
}

# Save it as a CSV file
df = pd.DataFrame(data)
df.to_csv('test.csv', index=False)
print("CSV created!")
```

```python
import pandas as pd

# Load the supermarket dataset
df = pd.read_csv('supermarket.csv', encoding='latin1')

# Display the first few rows
print(df.head())
```

```python
# Display column names and data types
print(df.info())

# Display summary statistics
print(df.describe())
```

```python
import matplotlib.pyplot as plt

# Bar chart of Total Sales by Category
df.groupby('Category')['Sales'].sum().plot(kind='bar', color='steelblue')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
